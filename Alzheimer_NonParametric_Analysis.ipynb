## 4. CONTRASTES DE BONDAD DE AJUSTE

La prueba chi cuadrado de bondad de ajuste es una prueba de hipótesis estadística que se usa para averiguar si es probable que una variable provenga o no de una distribución específica.

### 4.1. Prueba de bondad de ajuste sobre la variable `MemoryComplaints`

**Objetivo:**  
Comprobar si la proporción de personas que presentan quejas subjetivas de memoria (`MemoryComplaints`) se ajusta a una **distribución binomial teórica**, en la que se espera que el 50% de los individuos tenga quejas y el 50% no.

---

**Hipótesis:**

- **Hipótesis nula (H₀):**  
  La distribución de la variable `MemoryComplaints` sigue una distribución binomial discreta con dos posibles resultados (`0` = no presenta quejas, `1` = presenta quejas), con una probabilidad teórica de éxito $ p = 0.5 $.

- **Hipótesis alternativa (H₁):**  
  La distribución observada de `MemoryComplaints` **no** sigue una distribución binomial con parámetros $ n = 1 $, $ p = 0.5 $.

---

**Justificación del test:**  
La variable `MemoryComplaints` representa si un paciente refiere o no problemas subjetivos de memoria, una señal temprana que puede estar relacionada con deterioro cognitivo. Se trata de una variable categórica binaria, por lo que es adecuada para aplicar una prueba de bondad de ajuste utilizando el test Chi-cuadrado.

El contraste permite evaluar si la proporción observada de personas con quejas de memoria se ajusta a una proporción teórica del 50%, es decir, si ambos resultados (`0` y `1`) son igualmente probables en la población.

Además, este análisis sirve como punto de partida para explorar posteriormente si la variable `MemoryComplaints` está asociada con el diagnóstico de Alzheimer (`Diagnosis`) a través de un test de independencia.

#### 4.1.1. Análisis exploratorio DIRECTO de la variable MemoryComplaints

```python
# Contamos las frecuencias observadas de MemoryComplaints

tabla_freq = datos["MemoryComplaints"].value_counts().sort_index()
```

```python
tabla_freq
```

```python
# Extraemos frecuencias observadas como lista (ya en orden 0.0, 1.0)

observadas = tabla_freq.tolist()
```

```python
# Establecemos las etiquetas correspondientes

etiquetas = ['Sin quejas de memoria', 'Con quejas de memoria']   # 0 (no) y 1 (sí)
```

```python
# Gráfico de barras

plt.figure(figsize=(6, 4))
plt.bar(etiquetas, tabla_freq)
plt.title('Frecuencia de quejas de memoria')
plt.ylabel('Frecuencia')
plt.xlabel('Categoría')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculamos las frecuencias esperadas bajo la hipótesis de 50%-50%

total = sum(observadas)
esperadas = [total / 2, total / 2]
```

```python
# Aplicamos el contraste de bondad de ajuste (Chi-cuadrado)

chi2_stat, p_valor = chisquare(f_obs=observadas, f_exp=esperadas)
```

```python
# Presentamos los resultados

resultados = pd.DataFrame({
    'Categoría': etiquetas,
    'Frecuencia observada': observadas,
    'Frecuencia esperada': esperadas
})

resultados
```

```python
# Mostramos resultados estadísticos 

print("\n--- Resultados del contraste ---")
print(f"Estadístico Chi-cuadrado: {chi2_stat:.4f}")
print(f"p-valor: {p_valor:.2e}")
```

```python
# Interpretación del resultado del contraste

alpha = 0.05

if p_valor < alpha:
    display(Markdown("**Se rechaza la hipótesis nula (H₀).**"))

    print("La distribución observada de quejas de memoria difiere significativamente de una distribución binomial con n = 1 y p = 0.5.")
    print("Es decir, no es compatible con una proporción teórica del 50% de personas con y sin quejas de memoria.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad de los datos con la hipótesis nula (que asume una proporción del 50% de personas con y sin quejas de memoria) es del {:.2%}, lo que indica baja compatibilidad.".format(p_valor))

    print("\nConclusión 2:")
    print("Si la suposición de una proporción del 50% de personas con y sin quejas de memoria fuera cierta, la probabilidad de encontrar valores tan extremos como el observado sería de {:.2%}, lo cual es una probabilidad baja y poco atribuible al azar en este contexto.".format(p_valor))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I si rechazamos la hipótesis nula de una proporción 50%-50%. Dado que {:.2%} < 5%, rechazamos H₀ y concluimos que la proporción observada difiere significativamente del 50% teórico.".format(p_valor))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, podemos rechazar la hipótesis nula y afirmar que la distribución observada de quejas de memoria no sigue una proporción 50%-50%.")
else:
    display(Markdown("**No se rechaza la hipótesis nula (H₀).**"))

    print("La distribución observada de quejas de memoria es compatible con una distribución binomial con n = 1 y p = 0.5.")
    print("Es decir, con una proporción teórica del 50% de personas con y sin quejas de memoria.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad de los datos con la hipótesis nula (que asume una proporción del 50% de personas con y sin quejas de memoria) es del {:.2%}, lo que indica buena compatibilidad.".format(p_valor))

    print("\nConclusión 2:")
    print("Si la suposición de una proporción del 50% de personas con y sin quejas de memoria fuera cierta, la probabilidad de encontrar valores tan extremos como el observado sería de {:.2%}, lo cual es una probabilidad razonablemente alta, por lo que puede atribuirse al azar.".format(p_valor))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I si rechazamos la hipótesis nula de una proporción 50%-50%. Dado que {:.2%} > 5%, no la rechazamos y no hay evidencia suficiente para afirmar una desviación significativa.".format(p_valor))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, no podemos rechazar la hipótesis nula, por lo que consideramos que la distribución observada es compatible con una proporción del 50% de personas con y sin quejas de memoria.")
```

```python
# Test de bondad de ajuste (Chi-cuadrado)

chi2_stat, p_valor = chisquare(f_obs=observadas, f_exp=esperadas)
```

```python
# Presentamos los resultados

resultados = pd.DataFrame({
    'Categoría': etiquetas,
    'Frecuencia observada': observadas,
    'Frecuencia esperada': esperadas
})

resultados
```

### 4.3. Prueba de bondad de ajuste sobre la variable `Ethnicity`

**Objetivo:**  
Analizar si la distribución de los grupos étnicos (`Ethnicity`) en el conjunto de datos se ajusta a una **distribución teórica uniforme**, en la que se esperaría que todas las categorías tuvieran la misma representación.

**Motivo del análisis:**  
Tras haber aplicado la prueba de bondad de ajuste a otras variables categóricas ordinales, se propone ahora examinar una variable categórica nominal como es la etnia de los pacientes. Esta variable recoge la pertenencia étnica de cada individuo, codificada de la siguiente manera:

- 0 = Caucasian  
- 1 = African American  
- 2 = Asian  
- 3 = Other

Dado que esta variable representa diferentes grupos poblacionales, es relevante estudiar si el conjunto de datos presenta una distribución equitativa entre ellos o si, por el contrario, hay una sobrerrepresentación de algunos grupos respecto a otros. Este análisis puede ayudar a detectar posibles sesgos en la muestra o desigualdades estructurales en la representación étnica, lo cual es importante tanto en estudios médicos como sociales.

**Hipótesis:**

- **Hipótesis nula (H₀):**  
  La distribución de la variable `Ethnicity` es uniforme: todas las categorías (0 a 3) tienen la misma probabilidad de aparecer.

- **Hipótesis alternativa (H₁):**  
  La distribución observada difiere significativamente de una distribución uniforme.

**Justificación del test:**  
Como se comparan las frecuencias observadas con las frecuencias esperadas bajo una distribución uniforme (equiprobable), se aplica la prueba de bondad de ajuste Chi-cuadrado, que permite evaluar si las desviaciones entre ambas distribuciones son estadísticamente significativas o atribuibles al azar.

#### 4.2.1. Análisis exploratorio DIRECTO de la variable Ethnicity

```python
# Calculamos las frecuencias esperadas bajo la hipótesis de distribución uniforme

total = sum(observadas)
k = len(observadas)
esperadas = [total / k] * k     # misma frecuencia esperada para cada categoría
```

```python
# Aplicamos el contraste de bondad de ajuste

chi2_stat, p_valor = chisquare(f_obs=observadas, f_exp=esperadas)
```

```python
# Interpretación

alpha = 0.05

# Interpretación del resultado del contraste
if p_valor < alpha:
    display(Markdown("**Se rechaza la hipótesis nula (H₀).**"))

    print("La distribución observada de grupos étnicos difiere significativamente de una distribución uniforme.")
    print("Esto indica que algunos grupos están sobrerrepresentados o infrarrepresentados en la muestra.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad de los datos con la hipótesis nula (que plantea que todos los grupos étnicos tienen igual representación) es del {:.2%}, lo que indica baja compatibilidad.".format(p_valor))

    print("\nConclusión 2:")
    print("Si todos los grupos étnicos tuvieran igual representación, la probabilidad de encontrar una desviación tan extrema como la observada en las frecuencias de etnia sería de {:.2%}, lo cual es una probabilidad baja y poco atribuible al azar.".format(p_valor))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I si rechazamos la hipótesis de distribución uniforme. Como {:.2%} < 5%, rechazamos H₀ y concluimos que la representación étnica no es equitativa.".format(p_valor))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, podemos rechazar la hipótesis nula y afirmar que la distribución étnica observada no se ajusta a una distribución uniforme.")
else:
    display(Markdown("**No se rechaza la hipótesis nula (H₀).**"))

    print("No hay evidencia significativa para afirmar que la distribución observada de grupos étnicos difiere de una distribución uniforme.")
    print("Los datos muestran una representación relativamente equilibrada entre las distintas categorías étnicas.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad de los datos con la hipótesis nula (que plantea que todos los grupos étnicos tienen igual representación) es del {:.2%}, lo que indica buena compatibilidad.".format(p_valor))

    print("\nConclusión 2:")
    print("Si todos los grupos étnicos tuvieran igual representación, la probabilidad de encontrar una desviación tan extrema como la observada sería de {:.2%}, una probabilidad razonablemente alta que puede atribuirse al azar.".format(p_valor))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I si rechazamos la hipótesis de distribución uniforme. Como {:.2%} > 5%, no la rechazamos y no hay evidencia suficiente para afirmar lo contrario.".format(p_valor))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, no podemos rechazar la hipótesis nula, por lo que consideramos que la distribución observada de etnias es compatible con una distribución uniforme.")
```

```python
# Decisión por p-valor

if p_val < alpha:
    markdown_decision = Markdown("**Se rechaza la hipótesis nula (H₀).**")

    decision_pvalor = (
        "La distribución observada de los grupos étnicos difiere significativamente de una distribución uniforme. "
        "Esto sugiere que algunos grupos están sobrerrepresentados o infrarrepresentados en la muestra.\n\n"

        "Conclusión 1:\n"
        "El grado de compatibilidad de los datos con la hipótesis nula (que plantea igual proporción entre todos los grupos étnicos) "
        "es del {:.2%}, lo que indica baja compatibilidad.\n\n".format(p_val) +

        "Conclusión 2:\n"
        "Si todos los grupos étnicos tuvieran igual representación, la probabilidad de obtener una diferencia tan extrema como la observada sería de {:.2%}, "
        "lo cual es bajo y difícilmente atribuible al azar.\n\n".format(p_val) +

        "Conclusión 3:\n"
        "El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de distribución uniforme. "
        "Como {:.2%} < 5%, rechazamos H₀ y concluimos que la muestra presenta un reparto desigual entre los grupos étnicos.\n\n".format(p_val) +

        "Conclusión 4:\n"
        "Con un margen de error del 5%, podemos rechazar la hipótesis nula y afirmar que la representación étnica en la muestra no sigue una distribución uniforme."
    )
else:
    markdown_decision = Markdown("**No se rechaza la hipótesis nula (H₀).**")

    decision_pvalor = (
        "No hay evidencia suficiente para afirmar que la distribución de los grupos étnicos difiere significativamente de una distribución uniforme. "
        "La muestra parece mantener un reparto razonablemente equilibrado entre las categorías.\n\n"

        "Conclusión 1:\n"
        "El grado de compatibilidad de los datos con la hipótesis nula (que plantea igual proporción entre todos los grupos étnicos) "
        "es del {:.2%}, lo que indica buena compatibilidad.\n\n".format(p_val) +

        "Conclusión 2:\n"
        "Si todos los grupos étnicos tuvieran igual representación, la probabilidad de obtener una diferencia tan extrema como la observada sería de {:.2%}, "
        "lo cual es razonablemente alto y atribuible al azar.\n\n".format(p_val) +

        "Conclusión 3:\n"
        "El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de distribución uniforme. "
        "Como {:.2%} > 5%, no rechazamos H₀ y no hay evidencia suficiente para afirmar lo contrario.\n\n".format(p_val) +

        "Conclusión 4:\n"
        "Con un margen de error del 5%, no podemos rechazar la hipótesis nula, por lo que consideramos que la representación étnica observada es compatible con una distribución uniforme."
    )
```

```python
# Residuos estandarizados

residuos_estandarizados = ((o_i['count'] - e_i['proporcion']) / np.sqrt(e_i['proporcion'])).round(2)
```

```python
# Presentamos resultados en DataFrame

resultados = pd.DataFrame({
    'Categoría': etiquetas,
    'Frecuencia observada': o_i['count'].tolist(),
    'Frecuencia esperada': e_i['proporcion'].tolist(),
    'Residuo estandarizado': residuos_estandarizados.tolist()
})

resultados
```

```python
# Mostramos resultados del contraste

print("\n--- Resultados del contraste ---")
print(f"Estadístico Chi-cuadrado: {chi_0:.4f}")
print(f"Valor crítico (α = 0.05): {valor_critico:.4f}")

# Mostramos decisión 
display(markdown_decision)

# Conclusiones
print(f"\nConclusión (región crítica):\n{decision_region}")
print(f"--------------------------------------------------------------------------------------------------------------------------------------------------------")
print(f"\nConclusión (p-valor):\n{decision_pvalor}")
```

## 5. CONTRASTES DE HOMOGENEIDAD DE DOS MUESTRAS

**Justificación del test seleccionado:**  
Para comparar los valores de variables entre personas con y sin diagnóstico de Alzheimer (`Diagnosis`), se plantea un contraste no paramétrico de comparación de dos muestras independientes. En este caso, los grupos a comparar (con y sin diagnóstico) son completamente independientes entre sí.

El test seleccionado es el **Mann-Whitney U**, que permite evaluar si las distribuciones de una variable cuantitativa difieren significativamente entre dos grupos. Este contraste se basa en los rangos de los datos, no requiere suponer normalidad y es especialmente útil cuando los datos presentan asimetrías o valores atípicos, como suele ocurrir en medidas clínicas como `MMSE`.

### 5.1. Contraste U de Mann-Whitney para comparar la puntuación cognitiva (MMSE) según diagnóstico de Alzheimer

**Objetivo:**  
Evaluar si existen diferencias significativas en la puntuación del Mini-Mental State Examination (`MMSE`) entre personas con y sin diagnóstico de Alzheimer (`Diagnosis`) utilizando un contraste no paramétrico para muestras independientes.

**Contexto y motivo del análisis:**  
Tras haber analizado la distribución de variables categóricas mediante contrastes de bondad de ajuste e independencia, se propone ahora un contraste que permita comparar una variable cuantitativa continua (`MMSE`) entre dos grupos bien definidos: personas con y sin diagnóstico de Alzheimer.

La puntuación `MMSE` es una herramienta comúnmente utilizada en la evaluación del deterioro cognitivo. Dado que esta puntuación disminuye en presencia de enfermedades neurodegenerativas como el Alzheimer, resulta relevante estudiar si las distribuciones de `MMSE` difieren significativamente entre los dos grupos.

**Hipótesis:**

- **Hipótesis nula (H₀):**  
  Las distribuciones de `MMSE` en personas con y sin diagnóstico de Alzheimer son iguales.

- **Hipótesis alternativa (H₁):**  
  Las distribuciones de `MMSE` en personas con y sin diagnóstico de Alzheimer son diferentes.

```python
file_path = "alzheimers_disease_data.csv"
datos = pd.read_csv(file_path)
```

```python
# Separamos los datos por grupo de diagnóstico

mmse_con_alz = datos[datos['Diagnosis'] == 1]['MMSE']
mmse_sin_alz = datos[datos['Diagnosis'] == 0]['MMSE']
```

```python
# Estadísticos descriptivos

print("Resumen estadístico de la puntuación MMSE:")

print("\nPersonas con diagnóstico de Alzheimer:")
print(mmse_con_alz.describe())
print("\nPersonas sin diagnóstico de Alzheimer:")
print(mmse_sin_alz.describe())
```

```python
# Test de Mann-Whitney U

stat, p = mannwhitneyu(mmse_sin_alz, mmse_con_alz, alternative='two-sided')

# Mostramos resultados

print(f"\nMann-Whitney U: estadístico = {stat:.4f}, p-valor = {p:.2e}")
```

```python
# Interpretación

alpha = 0.05

if p < alpha:
    display(Markdown("**Se rechaza la hipótesis nula (H₀).**"))

    print("La diferencia observada en las puntuaciones MMSE entre personas con y sin diagnóstico de Alzheimer es estadísticamente significativa.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad de los datos con la hipótesis nula (que plantea igualdad en la distribución de puntuaciones MMSE entre ambos grupos) es del {:.2%}, lo que indica baja compatibilidad.".format(p))

    print("\nConclusión 2:")
    print("Si hay igualdad en la distribución de puntuaciones MMSE entre ambos grupos, la probabilidad de obtener una diferencia tan extrema como la observada sería de {:.2%}, lo cual es una probabilidad baja y difícilmente atribuible al azar.".format(p))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de igualdad de distribuciones. Como {:.2%} < 5%, rechazamos H₀ y concluimos que las puntuaciones MMSE difieren significativamente entre los grupos.".format(p))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, podemos rechazar la hipótesis nula y afirmar que el nivel de deterioro cognitivo medido por el test MMSE difiere entre personas con y sin Alzheimer.")
else:
    display(Markdown("**No se rechaza la hipótesis nula (H₀).**"))

    print("No se han encontrado diferencias estadísticamente significativas en las puntuaciones MMSE entre personas con y sin diagnóstico de Alzheimer.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad de los datos con la hipótesis nula (que plantea igualdad en la distribución de puntuaciones MMSE entre ambos grupos) es del {:.2%}, lo que indica buena compatibilidad.".format(p))

    print("\nConclusión 2:")
    print("Si hay igualdad en la distribución de puntuaciones MMSE entre ambos grupos, la probabilidad de obtener una diferencia tan extrema como la observada sería de {:.2%}, lo cual es razonablemente alto y atribuible al azar.".format(p))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de igualdad de distribuciones. Como {:.2%} > 5%, no rechazamos H₀ y no hay evidencia de que las puntuaciones MMSE difieran entre los grupos.".format(p))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, no podemos rechazar la hipótesis nula. Por tanto, no se puede afirmar que el nivel de deterioro cognitivo medido por el test MMSE sea diferente entre personas con y sin diagnóstico de Alzheimer.")
```

Los resultados gráficos lo refuerzan:

Para ello, aplicamos un **test de rachas** (*runs test*), un contraste no paramétrico que permite, entre otras cosas, analizar la aleatoriedad en una secuencia de valores binarios. En este caso:

- Ordenamos la variable binaria `Depression` en función de una variable continua `MMSE` (de menor a mayor puntuación).

- **Hipótesis alternativa (H₁):**

## 7. CONTRASTES DE NORMALIDAD

### 7.1. Contraste de Anderson-Darling

El test de Anderson-Darling compara si una muestra proviene de una distribución específica, siendo la normal la opción por defecto. A diferencia de otros tests, devuelve un estadístico y múltiples niveles de significación con sus respectivos valores críticos.

Para cada variable: si el estadístico es menor que el valor crítico para un nivel de significación (por ejemplo 5%), se considera que **no hay evidencia para rechazar la normalidad** a ese nivel.

Este test permite ver en qué niveles de significación una variable se comporta como normal o no.
Se comportan mal con muestras grandes, de varios miles de datos. En nuestro caso, el dataset tiene un poco más de 2000 muestras, que dentro de que son miles, no es un valor extremadamente alto, por lo que vamos a probar si el contraste funciona correctamente.

```python
# Lista para almacenar los resultados

anderson_results = []
```

```python
# Aplicamos el test de Anderson-Darling a cada variable numérica

for col in datos.select_dtypes(include='number').columns:
    result = anderson(datos[col].dropna())
    conclusions = []

    # Evaluamos en todos los niveles de significación disponibles
    for i in range(len(result.critical_values)):
        level = result.significance_level[i]
        crit_val = result.critical_values[i]

        # Comparación del estadístico con cada valor crítico
        if result.statistic < crit_val:
            conclusions.append(f"Normal al {level}%")
        else:
            conclusions.append(f"No normal al {level}%")

    # Guardamos resultados
    anderson_results.append((col, round(result.statistic, 4), "; ".join(conclusions)))
```

```python
# Mostramos resultados en DataFrame

df_anderson = pd.DataFrame(anderson_results, columns=["Variable", "Estadístico", "Conclusión"])

# Aumentamos el ancho de visualización de columnas
pd.set_option('display.max_colwidth', None)

# Mostramos el DataFrame completo con las conclusiones
display(df_anderson)
```

Ninguna variable se puede considerar normal según el contraste de Anderson-Darling.

### 7.2. Contraste de Shapiro-Wilk

Este test evalúa si una muestra proviene de una distribución normal. El test devuelve un estadístico y un p-valor:

- Si el p-valor > 0.05 → **no se rechaza la hipótesis de normalidad**.
- Si el p-valor ≤ 0.05 → **se rechaza la hipótesis de normalidad**.

A continuación, aplicamos el test a todas las variables numéricas del dataset para evaluar cuáles podrían aproximarse a una distribución normal.

```python
# Lista para almacenar los resultados

shapiro_results = []
```

```python
# Recorremos todas las variables numéricas del dataset

for col in datos.select_dtypes(include='number').columns:
    # Aplicamos el test de Shapiro-Wilk a cada variable (sin valores nulos)
    stat, p = shapiro(datos[col].dropna())

    # Conclusión basada en el p-valor
    conclusion = "Normal" if p > 0.05 else "No normal"

    # Guardamos los resultados
    shapiro_results.append((col, round(stat, 4), round(p, 4), conclusion))
```

```python
# Mostramos los resultados 

df_shapiro = pd.DataFrame(shapiro_results, columns=["Variable", "Estadístico", "p-valor", "Conclusión"])
display(df_shapiro)
```

Ninguna variable se puede considerar normal según el contraste de Shapiro-Wilk.

### 7.3. Test de D’Agostino-Pearson

Este test combina la asimetría (skewness) y la curtosis (kurtosis) de una variable para determinar si su distribución se desvía de la normalidad. 

- Si el p-valor > 0.05 → **la variable podría seguir una distribución normal**.
- Si el p-valor ≤ 0.05 → **se rechaza la hipótesis de normalidad**.

A continuación, se evalúan todas las variables numéricas del dataset mediante este test.

```python
# Lista para almacenar los resultados

dagostino_results = []
```

```python
# Aplicamos el test a cada variable numérica del dataset

for col in datos.select_dtypes(include='number').columns:
    stat, p = normaltest(datos[col].dropna())
    conclusion = "Normal" if p > 0.05 else "No normal"
    dagostino_results.append((col, round(stat, 4), round(p, 4), conclusion))
```

### 7.4. Conclusión de los contrastes de normalidad

El objetivo es mostrar las diferencias entre aplicar un test paramétrico y uno no paramétrico cuando los supuestos del primero (normalidad de los datos) no se cumplen. Aunque ambos tests evalúan si hay diferencias entre dos grupos, su fiabilidad depende de algunas condiciones.

**Variables seleccionadas**

- **Variable continua:** `BMI` (Índice de Masa Corporal), elegida por ser una medida biomédica relevante y continua.
- **Variable categórica binaria:** `Diagnosis` (0 = No Alzheimer, 1 = Alzheimer), utilizada para dividir la muestra en dos grupos clínicamente diferenciados.

**Supuestos**

- El test t de Student asume que las muestras provienen de distribuciones normales.
- El test U de Mann-Whitney es usado para muestras independientes de poblaciones no necesariamente normales.
- 
**Hipótesis de contraste**

Para ambos tests se plantea:

- **Hipótesis nula (H₀):** No hay diferencia en el BMI entre los grupos con y sin diagnóstico. Es decir, las muestras provienen de poblaciones que se distribuyen igual.
- **Hipótesis alternativa (H₁):** Sí hay diferencia en el BMI entre los grupos. Es decir, las muestras provienen de poblaciones que se distribuyen de forma distinta.

**¿Qué esperamos?**

Dado que las pruebas de normalidad indican que el BMI no se distribuye normalmente en ninguno de los dos grupos, el test t de Student no debería utilizarse en este caso. Lo que esperamos es que el test de Mann-Whitney proporcione una evaluación más fiable de las diferencias entre grupos, aunque el resultado final puede coincidir o no con el test t. ### 8.2. Coeficiente de correlación de Pearson vs Coeficiente de correlación de rangos de Spearman

La idea de este apartado es comparar dos métodos estadísticos para analizar la relación entre dos variables continuas: la edad (`Age`) de los pacientes y los niveles de colesterol HDL (`CholesterolHDL`).

- **Correlación de Pearson** (método paramétrico)  
- **Correlación de Spearman** (método no paramétrico)

**Objetivo**

El objetivo es mostrar las diferencias entre aplicar un método paramétrico y uno no paramétrico cuando los supuestos del primero (normalidad de los datos) no se cumplen. Ambos métodos evalúan si existe una asociación entre dos variables, pero difieren en el tipo de relación que detectan y en los requisitos sobre la distribución de los datos.

**Variables seleccionadas**

- **Variable 1:** `Age` (edad), variable numérica continua que representa la edad del paciente.
- **Variable 2:** `CholesterolHDL` (colesterol HDL), variable numérica continua asociada al perfil lipídico del paciente, que mide el nivel de colesterol HDL.

**Supuestos**

- La correlación de Pearson evalúa relaciones lineales y asume que ambas variables siguen una distribución normal.
- La correlación de Spearman evalúa relaciones monótonas y no requiere normalidad.

**Hipótesis de contraste**

Para ambas correlaciones se plantea:

- **Hipótesis nula (H₀):** No hay asociación significativa entre Age y CholesterolHDL.
- **Hipótesis alternativa (H₁):** Sí hay una asociación significativa entre Age y CholesterolHDL.

**¿Qué esperamos?**

Dado que las pruebas de normalidad muestran que ni `CholesterolHDL` ni `Age` se distribuyen normalmente, la correlación de Pearson no es apropiada en este contexto. Se espera que la correlación de Spearman proporcione una evaluación más robusta de la posible relación. 

```python
# Seleccionamos las variables continuas y eliminamos nulos

df_corr = datos[['Age', 'CholesterolHDL']].dropna()
x = df_corr['Age']
y = df_corr['CholesterolHDL']
```

```python
# Verificamos la normalidad con D’Agostino 

normal_age = normaltest(x)
normal_hdl = normaltest(y)

print(f"Normalidad Age: p-value = {normal_age.pvalue:.3e}")
print(f"Normalidad CholesterolHDL: p-value = {normal_hdl.pvalue:.3e}")
```

```python
if normal_age.pvalue < 0.05 or normal_hdl.pvalue < 0.05:
    print("Alguna de las variables no sigue una distribución normal (p-valor < 0.05). Por lo tanto, el coeficiente de correlación de Pearson no es apropiado.")
else:
    print("Ambas variables parecen normales (p-valor ≥ 0.05), se puede usar el coeficiente de correlación de Pearson.")
```

```python
# Calculamos correlacion de Pearson (paramétrico)

pearson_corr = pearsonr(x, y)

print(f"\nCorrelación de Pearson: r = {pearson_corr.statistic:.4f}, p-value = {pearson_corr.pvalue:.3f}")
```

```python
# Interpretación del p-valor para la correlación de Pearson (paramétrica)

if pearson_corr.pvalue < 0.05:
    display(Markdown("**Se rechaza la hipótesis nula (H₀) en la correlación de Pearson.**"))

    print("La correlación de Pearson indica una relación significativa entre Age y CholesterolHDL.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad con la hipótesis nula (que plantea ausencia de correlación lineal) es del {:.2%}, lo que indica baja compatibilidad.".format(pearson_corr.pvalue))

    print("\nConclusión 2:")
    print("Si la ausencia de correlación lineal fuera cierta, la probabilidad de obtener una correlación tan extrema como la observada sería de {:.2%}, lo cual es poco atribuible al azar.".format(pearson_corr.pvalue))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de ausencia de correlación. Como {:.2%} < 5%, rechazamos H₀ y concluimos que hay relación lineal entre las variables.".format(pearson_corr.pvalue))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, podemos rechazar la hipótesis nula y afirmar que existe una relación lineal entre la edad y el colesterol HDL, bajo el supuesto de normalidad.")
else:
    display(Markdown("**No se rechaza la hipótesis nula (H₀) en la correlación de Pearson.**"))

    print("La correlación de Pearson no encuentra una relación significativa entre Age y CholesterolHDL.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad con la hipótesis nula (que plantea ausencia de correlación lineal) es del {:.2%}, lo que indica buena compatibilidad.".format(pearson_corr.pvalue))

    print("\nConclusión 2:")
    print("Si la ausencia de correlación lineal fuera cierta, la probabilidad de obtener una correlación tan extrema como la observada sería de {:.2%}, una probabilidad atribuible al azar.".format(pearson_corr.pvalue))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de ausencia de correlación. Como {:.2%} > 5%, no la rechazamos.".format(pearson_corr.pvalue))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, no podemos rechazar la hipótesis nula, por lo que no hay evidencia estadística de que exista una relación lineal entre la edad y el colesterol HDL.")
```

```python
# Calculamos correlacion de Spearman (no paramétrico)

spearman_corr = spearmanr(x, y)

print(f"Correlación de Spearman: ρ = {spearman_corr.correlation:.4f}, p-value = {spearman_corr.pvalue:.3f}")
```

```python
# Interpretación del p-valor para la correlación de Spearman (no paramétrica)

if spearman_corr.pvalue < 0.05:
    display(Markdown("**Se rechaza la hipótesis nula (H₀) en la correlación de Spearman.**"))

    print("La correlación de Spearman indica una relación significativa entre Age y CholesterolHDL.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad con la hipótesis nula (que plantea ausencia de correlación monótona) es del {:.2%}, lo que indica baja compatibilidad.".format(spearman_corr.pvalue))

    print("\nConclusión 2:")
    print("Si la ausencia de correlación monótona fuera cierta, la probabilidad de obtener una correlación tan extrema como la observada sería de {:.2%}, lo cual es poco atribuible al azar.".format(spearman_corr.pvalue))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de independencia. Como {:.2%} < 5%, rechazamos H₀ y concluimos que existe una asociación monótona entre las variables.".format(spearman_corr.pvalue))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, podemos rechazar la hipótesis nula y afirmar que existe una relación monótona entre la edad y el colesterol HDL, sin necesidad de asumir normalidad.")
else:
    display(Markdown("**No se rechaza la hipótesis nula (H₀) en la correlación de Spearman.**"))

    print("La correlación de Spearman no encuentra una relación significativa entre Age y CholesterolHDL.\n")

    print("Conclusión 1:")
    print("El grado de compatibilidad con la hipótesis nula (que plantea ausencia de correlación monótona) es del {:.2%}, lo que indica buena compatibilidad.".format(spearman_corr.pvalue))

    print("\nConclusión 2:")
    print("Si la ausencia de correlación monótona fuera cierta, la probabilidad de obtener una correlación tan extrema como la observada sería de {:.2%}, lo cual es razonablemente alto y atribuible al azar.".format(spearman_corr.pvalue))

    print("\nConclusión 3:")
    print("El p-valor representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis de independencia. Como {:.2%} > 5%, no la rechazamos.".format(spearman_corr.pvalue))

    print("\nConclusión 4:")
    print("Con un margen de error del 5%, no podemos rechazar la hipótesis nula, por lo que no hay evidencia estadística de relación entre la edad y el colesterol HDL desde un enfoque no paramétrico.")
```

```python
# Visualización
plt.figure(figsize=(7, 6))
sns.scatterplot(x=x, y=y, alpha=0.4)
sns.regplot(x=x, y=y, scatter=False, color='red', label=f'Ajuste lineal (Pearson r = {pearson_corr[0]:.4f})')

# Etiquetas y título
plt.xlabel("Edad")
plt.ylabel("Colesterol HDL")
plt.title(f"Relación entre Edad y Colesterol HDL\nSpearman ρ = {spearman_corr.correlation:.4f}, p-value = {spearman_corr.pvalue:.3f}")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Esto indica que no hay evidencia suficiente para afirmar que exista una relación lineal ni monótona entre ambas variables.

Sin embargo, dado que las pruebas de normalidad mostraron que al menos una de las dos variables no sigue una distribución normal, la correlación de Spearman es más adecuada para este análisis. Aunque el resultado sea similar al de Pearson, es importante resaltar que la validez del método paramétrico puede verse afectada cuando no se cumplen sus supuestos, lo que justifica la preferencia por métodos no paramétricos en este tipo de situaciones.

Además, se observa que el p-valor de Spearman es ligeramente menor que el de Pearson, lo cual puede interpretarse como una mayor sensibilidad del método no paramétrico en contextos donde los datos no se distribuyen normalmente, es decir, una mayor capacidad para detectar una posible asociación cuando realmente existe.
